```python
import numpy as np
import pandas as pd
```

```python
header = ['user_id', 'item_id', 'rating', 'timestamp']
df_data = pd.read_csv('ml-100k/u.data', sep='\t', names=header)
df_data
```

```python
header = ['user_id', 'age', 'Gender', 'Occupation', 'zip_code']
df_user = pd.read_csv('ml-100k/u.user', sep='|', names=header)
df_user
```

```python
df_data = df_data[df_data['rating'] == 5]
df_data
```

```python
df_movie = pd.read_csv('ml-100k/u.item', sep="|", encoding="iso-8859-1")
df_movie = df_movie.iloc[:, 0:2]
df_movie.columns = ['item_id', 'name']
df_movie
```

```python
df = pd.merge(df_user, df_data, on='user_id', how='inner')
df = pd.merge(df_movie, df, on='item_id', how='inner')
df
```

```python
male_df = df[df['Gender'] == 'M']
female_df = df[df['Gender'] == 'F']
print(len(male_df))
print(len(female_df))
```

```python
grp_male = male_df.groupby(['Gender', 'name']).agg({'rating':'count'}).sort_values(by=['rating'], ascending=False)
grp_male.head(10)
```

```python
grp_female = female_df.groupby(['Gender', 'name']).agg({'rating':'count'}).sort_values(by=['rating'], ascending=False)
grp_female.head(10)
```

```python
df.groupby('Occupation').agg({'rating':'count'}).sort_values(by=['rating'], ascending=False)
```

```python
grpd_fe = female_df.groupby(['Gender', 'Occupation']).agg({'rating':'count'}).sort_values(by=['rating'], ascending=False)
```

```python
grpd_ma = male_df.groupby(['Gender', 'Occupation']).agg({'rating':'count'}).sort_values(by=['rating'], ascending=False)
```

```python
df_new = pd.merge(grpd_fe, grpd_ma, on='Occupation', how='inner')
df_new.columns = ['female users', 'male users']
df_new
```
